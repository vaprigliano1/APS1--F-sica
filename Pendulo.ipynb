```python
import math
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt


t = np.arange(0,5, 0.01)
g=9.8 #(m/s**2)
r=2


#---------------------------------------------- 10 graus


def EqDif(Y,t):

    teta = Y[0]
    omega = Y[1]
    
    dtetadt = omega
    domegadt = (-g*math.sin(teta))/r #aceleração tangencial
    
    return (dtetadt, domegadt)


S0=[math.radians(10),0]

Solucao=odeint(EqDif, S0, t)

plt.axis('equal')
plt.grid(True)
plt.plot(t, Solucao[:,0])
plt.show()

```

```python
#---------------------------------------------- 30 graus
S0=[math.radians(30),0]

Solucao=odeint(EqDif, S0, t)

plt.axis('equal')
plt.grid(True)
plt.plot(t, Solucao[:,0])
plt.show()

```

```python
#---------------------------------------------- 60 graus
S0=[math.radians(60),0]

Solucao=odeint(EqDif, S0, t)

plt.axis('equal')
plt.grid(True)
plt.plot(t, Solucao[:,0])
plt.show()
```

```python
#movimento do pêndulo

X=[]
Y=[]

for teta in Solucao[:,1]:
    X.append(r*math.cos(teta))
    
for teta in Solucao[:,0]:
    Y.append(r*math.sin(teta))

    
plt.grid(True)
plt.plot(Y,X)
plt.show()
```
